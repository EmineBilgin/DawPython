 # What country has the highest percentage of people that earn >50K?
 ---


```python
country_salary_df = df.groupby('native-country')[['salary']].count()
country_salary_df
```

```python
highsalary= df[df['salary'] == ">50K"]
highsalary
```

```python
country_salary_df_riches = highsalary.groupby('native-country')[['salary']].count()
country_salary_df_riches
```

```python
country_salary_df_riches.sum()
```

```python
rich_per_country =country_salary_df_riches/country_salary_df
df1 = rich_per_country.sort_values(by='salary',ascending=False)
df1
```

```python
df1.head(1)
```

```python
#use iloc here

df1.iloc[:1]
```
